```python
#1. What is a random variable in probability theory? 
#a random variable is a rule or function that assigns a numerical value to each possible outcome of a random experiment.
```

```python
#2. What are the types of random variables? 
#Types:Discrete random variable: Takes a countable number of possible values (e.g., number of sixes in 10 dice rolls).
#Continuous random variable: Takes values from an interval of real numbers (e.g., the exact height of a randomly chosen person).
```

```python
#3.Explain the difference between discrete and continuous distributions.
#In probability theory, a discrete distribution describes a situation where the random variable can take only countable values, such as the number of heads in coin tosses or the result of rolling a die.Each value has a specific probability greater than zero, represented by a probability mass function (PMF), and the total of all probabilities equals 1.
#a continuous distribution applies when the random variable can take any value within an interval, such as height, weight, or time. Here, the probability of any exact value is zero, and probabilities are assigned to ranges of values using a probability density function (PDF), where the area under the curve over a range gives the probability. The key difference is that discrete variables deal with separate, countable outcomes, while continuous variables cover an unbroken range of possibilities.
```

```python
#4.What is a binomial distribution, and how is it used in probability? 
#A binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials,
#where each trial has only two possible outcomes — usually called success and failure — and the probability of success remains constant in each trial.
```

```python
#5.What is the standard normal distribution, and why is it important? 
#The standard normal distribution is a special case of the normal (Gaussian) distribution where the mean is 0 and the standard deviation is 1.
#importantance:Universal reference,Probability calculations,Statistical inference
```

```python
#6.What is the Central Limit Theorem (CLT), and why is it critical in statistics? 
#The Central Limit Theorem says that the distribution of sample means approaches a normal distribution as the sample size grows, no matter the population’s shape. 
#It’s crucial because it allows statisticians to use normal-based methods for inference, even with non-normal data.
```

```python
#7.What is the significance of confidence intervals in statistical analysis? 
#A confidence interval provides a range of values within which the true population parameter is likely to fall, along with a specified confidence level (e.g., 95%). 
#It is significant in statistical analysis because it quantifies the uncertainty in estimates, offering more information than a single point estimate and helping to make informed, data-driven decisions.
```

```python
#8.What is the concept of expected value in a probability distribution?
#The expected value of a probability distribution is the long-run average or mean outcome you would expect if an experiment were repeated many times. It is calculated by multiplying each possible value of the random variable by its probability and then summing (for discrete distributions) or integrating (for continuous distributions) these products.
```

```python
#9.Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution. 
import numpy as np
import matplotlib.pyplot as plt
mean = 50
std_dev = 5
n_samples = 1000

# Generate random numbers from normal distribution
data = np.random.normal(mean, std_dev, n_samples)

# Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std = np.std(data)

print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std:.2f}")

# Plot histogram
plt.hist(data, bins=30, edgecolor='blue', alpha=0.7)
plt.title("Histogram ")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()
```

```python
#10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. 
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260] 
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales)
print(f"Average Daily Sales: {mean_sales:.2f}")
print(f"Standard Deviation: {std_sales:.2f}")
plt.plot(daily_sales, marker='o', linestyle='-', color='black', label='Daily Sales')
plt.axhline(y=mean_sales, color='pink', linestyle='--', label=f'Average Sales ({mean_sales:.2f})')
plt.title("Daily Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
